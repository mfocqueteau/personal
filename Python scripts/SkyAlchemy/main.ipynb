```python
import pandas as pd
import numpy as np
```

```python
# usecols=['Ingredient', 'Effect 1', 'Effect 2', 'Effect 3', 'Effect 4']
data = pd.read_csv('data.csv', delimiter='\t')
data.columns = ['Ingredient', 'Effect 1', 'Effect 2', 'Effect 3', 'Effect 4']
data.set_index('Ingredient', inplace=True)
data
```

```python
def has_any_effect(*effects):
    return data[data.isin(effects).any(axis=1)]


def has_all_effect(effect: str, *effects: str):
    result = data[data.isin([effect]).any(axis=1)]
    for e in effects:
        result = result[result.isin([e]).any(axis=1)]
    return result


def show_split(effect: str, *effects: str):
    if effects:
        print(
            'Ingredientes con todos los efectos:\n - ',
            '\n - '.join(has_all_effect(effect, *effects).index)
        )
    return has_any_effect(effect, *effects)
```

```python
uniques = []
for i in range(1, 5):
    uniques += list(data[f'Effect {i}'].unique())
uniques = tuple(set(uniques))
```

```python
array = np.zeros((len(data.index), len(data.index)), int)
for ing1, row1 in enumerate(data.values):
    for ing2, row2 in enumerate(data.values):
        if ing1 == ing2:
            array[ing1, ing2] = -1
            continue
        array[ing1, ing2] = len(tuple(e for e in row1 if e in row2))


df = pd.DataFrame(array, columns=data.index, index=data.index)
```

```python
def mix_with(ingredient: str):
    return pd.DataFrame(df[ingredient][df[ingredient] > 0]).sort_values(ingredient, ascending=False)
```
